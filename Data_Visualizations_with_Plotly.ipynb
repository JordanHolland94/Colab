```python
import pandas as pd
import plotly.express as px
```

```python
afc_standings = pd.read_html('https://www.pro-football-reference.com/years/2021/')[0]
afc_standings = afc_standings[afc_standings.columns[:8]]
afc_standings.columns = ['Team', 'Wins', 'Losses', 'Ties', 'WinLossPerc', 'PointsFor', 'PointsAgainst', 'PointDifferential']
afc_standings = afc_standings[~afc_standings.index.isin(list(range(0, len(afc_standings), 5)))]
afc_standings.loc[:, 'Team'] = afc_standings['Team'].str.replace('[^a-zA-Z0-9 ]', '', regex = True)  # Regex portion - ^ means DON'T match a-z, A-Z, 0-9, or spaces. Everything else gets returned and replaced.

afc_division_mapping = {
    'Baltimore Ravens': 'AFC North',
    'Buffalo Bills': 'AFC East',
    'Cincinnati Bengals': 'AFC North',
    'Cleveland Browns': 'AFC North',
    'Denver Broncos': 'AFC West',
    'Houston Texans': 'AFC South',
    'Indianapolis Colts': 'AFC South',
    'Jacksonville Jaguars': 'AFC South',
    'Kansas City Chiefs': 'AFC West',
    'Las Vegas Raiders': 'AFC West',
    'Los Angeles Chargers': 'AFC West',
    'Miami Dolphins': 'AFC East',
    'New England Patriots': 'AFC East',
    'New York Jets': 'AFC East',
    'Pittsburgh Steelers': 'AFC North',
    'Tennessee Titans': 'AFC South'
}

afc_standings['Division'] = afc_standings['Team'].map(afc_division_mapping)
```

```python
aes = afc_standings.query("Division == 'AFC East'")  # AFC East Standings (aes)
```

```python
fig = px.bar(data_frame = aes.sort_values('Wins', ascending = True),
             x = 'Team',
             y = 'Wins',
             title = 'Wins by Team')
fig.show()
```

```python
print(fig.layout)
```
